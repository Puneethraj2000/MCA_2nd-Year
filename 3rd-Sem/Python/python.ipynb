```python
#3.Write a python program using object oriented programming to demonstrate encapsulation,overloading and inheritance

#Encapsulation

class Students:
    def __init__(self, name, rank, points):
        self.name = name
        self.rank = rank
        self.points = points
        
    def printfunction(self):
        print("I am",self.name)
        print("I got rank",self.rank)
        
str1 = Students("Kiran",1,100)
str2 = Students("Indu",2,90)

str1.printfunction()
str2.printfunction()

#Inheritence

class parent:
    def func1(self):
        print("Hello Parent")

class child(parent):
    def func2(self):
        print("Hello Child")
        
test = child()
test.func1()
test.func2()

#Overloading

class Addition:
    def my_sum(self, a = None, b = None, c = None):
        s = 0
        if a != None and b != None and c != None:
            s = a + b + c
        elif a != None and b != None:
            s = a + b
        return s

obj = Addition()
print(obj.my_sum(10,20))
print(obj.my_sum(20,10,30))
print(obj.my_sum(20))
```

```python
from multipledispatch import  dispatch

@dispatch(int, int)
def mul(a, b):
    return a * b

@dispatch(int, int, int)
def mul(a, b, c):
    return a * b * c

@dispatch(float, float, float)
def mul(a, b, c):
    return a * b * c

print(mul(2.1, 4.5, 5.4))
print(mul(2, 4, 6))
print(mul(3, 5))
```

```python
#9.Write a Python program to demonstrate the generation of logistic regression models using Python.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

dataset = pd.read_csv("/home/mca/Puneethraj/DAP Datasets/User_Data.csv")
# input
x = dataset.iloc[:, [2, 3]].values

# output
y = dataset.iloc[:, 4].values

from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.25, random_state = 0)

from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
xtrain = sc_x.fit_transform(xtrain)
xtest = sc_x.transform(xtest)

print (xtrain[0:10, :])

#Here once see that Age and Estimated salary features values are scaled and now there in the -1 to 1. Hence, each feature will contribute equally to decision making i.e. finalizing the hypothesis.

from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(xtrain, ytrain)

y_pred = classifier.predict(xtest)

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(ytest, y_pred)
print ("Confusion Matrix : \n", cm)


from sklearn.metrics import accuracy_score
print ("Accuracy : ", accuracy_score(ytest, y_pred))


from matplotlib.colors import ListedColormap
X_set, y_set = xtest, ytest
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1,
                               stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1,
                               stop = X_set[:, 1].max() + 1, step = 0.01))

plt.contourf(X1, X2, classifier.predict(
             np.array([X1.ravel(), X2.ravel()]).T).reshape(
             X1.shape), alpha = 0.75, cmap = ListedColormap(('white', 'black')))

plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())

for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('red', 'black'))(i), label = j)

plt.title('Classifier (Test set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

```python
#10.Write a Python program to demonstrate Timeseries analysis with Pandas.
#Data can be in chronological. That means yearly monthly...
#Data is depend on the time.
#Quarter = Every 3 months
#Rule of Thumb = Pay and eat

import pandas as pd 
df = pd.read_csv("/home/mca/Puneethraj/DAP Datasets/AirPassengers.csv")
print(df.head())
print(df.tail())
df['Month'] = pd.to_datetime(df['Month'], format='%Y-%m')
print(df.head())
df.index = df['Month']
del df['Month']
print(df.head())

#Rolling Data = Data will have the irregular frequency. So we take the particular time period and calculate 
#               the average of that

import matplotlib.pyplot as plt
import seaborn as sns #extension of matplotlib

sns.lineplot(data = df)
plt.ylabel("Number of Passengers")

rolling_mean = df.rolling(7).mean()
rolling_std = df.rolling(7).std()

plt.plot(df, color = "blue",label = "Original Passenger Data")

plt.plot(rolling_mean, color = "red", label = "Rolling Mean Passenger Number")

plt.plot(rolling_std, color = "black", label = "Rolling Standard Deviation in Passenger Number")

plt.title("Passenger Time Series, Rolling Mean, Standard Deviation")

plt.legend(loc = "best")

from statsmodels.tsa.seasonal import seasonal_decompose
decompose = seasonal_decompose(df['#Passengers'], model = 'additive', period = 7)
decompose.plot()
plt.show()
```
